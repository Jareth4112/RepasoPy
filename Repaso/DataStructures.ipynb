## **DICTIONARIES**
- Similares a las listas pero con una clave y un valor.

````python
cab = dict()#empty dict
cab['summer'] = 3
cab['fall'] = 6
cab['spring'] = 9
print(cab['summer'])
````

````python
cab['fall'] = cab['spring'] + 2
print(cab)
````

````python
'winter' in cab
````

````python
c = dict()
names = ['csev', 'cwen', 'cesv', 'zqian', 'cwen']
for name in names:
    c[name] = c.get(name, 0) + 1
print(c)
````

````python
pdf1 = r'c:\Users\jaret\Desktop\Portafolio\Python\Extraer_datos\PDFs\MapfreSABMEMAP_PublicCompany.pdf'
doc = pdf.open(pdf1)
texto_completo = ""
for pagina in doc:
    texto_completo += pagina.get_text()

counts = dict()
pdf_split = texto_completo.split()
for lines in pdf_split:
    counts[lines] = counts.get(lines, 0) + 1
print(counts)
````

````python
for key in counts:
    print(key, counts[key])
````

````python
#A list of the keys:
print(list(counts.keys()))
````

````python
#A list of the value:
print(list(counts.values()))
````

````python
#A list of tuples of the values and keys of the dictionary
print(list(counts.items()))
````

````python
#Iteration between tuples
for aaa, bbb in counts.items():
    print(aaa, bbb)
````

````python
#Feel the word that repeats more times
big_count = None
big_word = None
for word, count in counts.items():
    if big_count is None or count > big_count:
        big_word = word
        big_count = count
print(big_word, big_count)
````

**Recurda que el metodo .item() de un diccionario nos devuelve una tupla de sus elementos**

````python
#A list of tuples of the values and keys of the dictionary
print(list(counts.items()))
````

````python
#Operador de comparacion, comienza por el primer elemento si es que hay diferencia
(0, 1, 2) < (5, 4, 2)
````

````python
('Jones', 'Sally') < ('Jones', 'Sam')
````

````python
#Podemos tomar ventaja de ordenar las tuplas para ordenar un  diccionario de acuerdo a la key
sorted(counts.items())
````

````python
for k, v in sorted(counts.items()):
    print(k, v)
````

````python
#Si queremos ordenar de acuerdo al valor, convertimos los valores en tuplas
tmp = list()
for v, k in counts.items():
    tmp.append((k, v))#Aca creo la tupla (k, v) -> tuplas 
tmp = sorted(tmp, reverse=True)
print(tmp)
````

````python
for v, k in tmp:
    print(v, k)
````

## ***LIST COMPREHENSION***
- Crea una lista a partir de otra lista, de un string, etc. de una manera mas sencilla y rapida.

```python

````python
lcomp = sorted([ (v, k) for k, v in counts.items()], reverse=True)
print(sorted([ (v, k) for k, v in counts.items()], reverse=True))

for v, k in lcomp:
    print(v, k)
````
